```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la distribución lognormal
media = 10  # Tiempo promedio de espera en minutos
desviacion_estandar = 2  # Desviación estándar en minutos

# Generar tiempos de espera para 50 clientes
n_clientes = 50
tiempos_espera = np.random.lognormal(mean=np.log(media), sigma=desviacion_estandar, size=n_clientes)

# Mostrar los resultados
print("Tiempos de espera para 50 clientes en el mostrador de servicio (en minutos):", tiempos_espera)

# Graficar histograma de los tiempos de espera
plt.hist(tiempos_espera, bins=20, edgecolor='black', density=True)
plt.title("Distribución de tiempos de espera en el mostrador de servicio")
plt.xlabel("Tiempo de espera (minutos)")
plt.ylabel("Frecuencia")
plt.show()
```
