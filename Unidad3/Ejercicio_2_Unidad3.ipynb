```python
import numpy as np

# Parámetro de la distribución exponencial (tasa de llegada)
tasa_llegada = 1 / 15  # 1 llegada cada 15 minutos

# Duración del día de operación en minutos (12 horas)
duracion_dia = 12 * 60

# Generar tiempos entre llegadas de vehículos eléctricos durante un día de operación
tiempos_entre_llegadas = np.random.exponential(scale=1/tasa_llegada, size=duracion_dia)

# Mostrar los resultados
print("Tiempos entre llegadas de vehículos eléctricos durante un día de operación (en minutos):", tiempos_entre_llegadas)
```
